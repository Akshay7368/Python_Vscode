```python
import pandas as pd

# Define logic functions
def get_total(row):
    return row['M1'] + row['M2'] + row['M3']

def get_avg(row):
    return row['Total'] / 3

def get_result(avg):
    return 'Pass' if avg >= 35 else 'Fail'

def get_grade(avg):
    if avg >= 75:
        return 'A'
    elif avg >= 60:
        return 'B'
    elif avg >= 35:
        return 'C'
    else:
        return 'F'

# Input data
data = {
    'Name': ['Akshay', 'Kumar', 'Meena'],
    'M1': [80, 30, 60],
    'M2': [75, 25, 50],
    'M3': [90, 40, 70]
}

df = pd.DataFrame(data)

# Apply logic cleanly
df['Total'] = df.apply(get_total, axis=1)
df['Avg'] = df.apply(get_avg, axis=1)
df['Result'] = df['Avg'].apply(get_result)
df['Grade'] = df['Avg'].apply(get_grade)

print(df)
```

```python
df = pd.DataFrame({'Category': np.random.choice(['A', 'B', 'C'], size=10), 'Value': np.random.randint(50, 100, size=10)})
print("Original DataFrame:")
print(df)

# Group the DataFrame by 'Category' and compute the sum and mean of 'Value' for each category
grouped = df.groupby('Category')['Value'].agg(['sum', 'mean'])
print("Grouped DataFrame:")
print(grouped)
```
