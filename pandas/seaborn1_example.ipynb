```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
 
data = {
    'Age': np.random.randint(20,45, size=15),
    'Salary': np.random.randint(10000, 40000, size=15),
    'department': np.random.choice(['Hr','Tl','Manager','Qc'], size=15)
    
}

df= pd.DataFrame(data)
plt.figure(figsize=(10,6))
plt.title("My team")
sns.boxplot(x='department', y= 'Salary', data=df)
sns
```

```python
salary_sum = df.groupby('department')['Salary'].sum()

plt.pie(salary_sum, labels=salary_sum.index, autopct='%1.1f%%', startangle=90)
plt.title('Total Salary Contribution by Department')
plt.show()
```
