```python
import numpy as np
import random 
import numpy.ma as ma
np.random.seed(5)
a = np.random.randint(1,30,size=(3,3))
b = np.random.randint(40,90, size=(3,3))
print("1st arary:\n",a)
print("2nd array:\n",b)
sum = a + b
corners = sum[[0,0,-1,-1],[0,-1,-1,0]]
msk_arr = ma.masked_where(sum < 70, sum)
msk_filled = ma.filled(msk_arr, 0)
print("sum:\n",sum)
print("corner elements:\n",corners)
print("masked array:\n",msk_arr)
print("msk filled:\n",msk_filled)
```

```python
Accounts = {'Akshay': 4000, 'Kumar': 3000, 'Jack': 8000, 'Peter':1000, 'Ben': 7400}

def Transfer(Sender, Receiver, Amount):
    if Sender not in Accounts or Receiver not in Accounts:
        print(f"{Sender}/{Receiver} is not in Accounts. Check it please! ")
    elif Amount <=0:
        print("Insufficient balance")
    elif Accounts[Sender] >= Amount:
        Accounts[Sender]-= Amount
        Accounts[Receiver]+=Amount
        print(f"The Amount is Successfully transfered from {Sender} to {Receiver}")
        print(f"New balance of {Sender}: {Accounts[Sender]}")
        print(f"New balance of {Receiver}: {Accounts[Receiver]}")
    else:
        print(f"{Sender} doesn't have enough balance to transfer ₹{Amount}")

i = 1
while i <= 5:
    print(f"\n--- Transaction {i} ---")
    sender = input("Enter the Sender Name: ")
    receiver = input("Enter the Receiver Name: ")
    amount = int(input("Enter the Amount: "))

    Transfer(Sender=sender, Receiver=receiver, Amount=amount)
    print("Transfer Completed")
    print("Next transaction")
    i += 1
print("All 5 transactions is completed")       
    
```
